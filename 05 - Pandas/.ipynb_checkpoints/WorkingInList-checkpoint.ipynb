```python
from csv import reader

opened_file = open("artworks.csv", encoding = "utf-8")
read_file = reader(opened_file)
lst = list(read_file)
lst_head = lst[0]
lst = lst[1:]
```

# Cleaning Null Values of Nationality and Gender Column

```python
def clean_and_convert(value, type):
    if value != "":
        #print("A") if a > b else print("B")
        #if value.count("(") > 0: value = value.replace("(","") 
        #if value.count(")") > 0: value = value.replace(")","") 
        value = value.replace("(","")
        value = value.replace(")","")
        value = value.replace("-","")
        value = value.replace("-","")
        if type == 'i':
            value = int(value)
        else:
            if value == "":
                value = "N/A"   
        return value
```

```python
for x in lst: 
    x[2] = clean_and_convert(x[2], 's')    
    x[3] = clean_and_convert(x[3], 'i')
    x[4] = clean_and_convert(x[4], 'i')
    x[5] = clean_and_convert(x[5], 's')  
```

```python
# Log Way to Clear the Data
for x in lst:
    #print(i, x[2], ' ', x[5])
    national = x[2]
    national = national.replace('(', '')
    national = national.replace(')', '')
    x[2] = national    
    
    age = x[5]
    age = age.replace('(', '')
    age = age.replace(')', '')
    x[5] = age  
    
for x in lst:
    a, b = (x[2], x[5])
    if a == "":
        a = 'N/A'
    if b == "":
        b = 'N/A'
    x[2] = a
    x[5] = b
```
