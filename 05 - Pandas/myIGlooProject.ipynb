```python
import numpy as np
import pandas as pd
import os
df = pd.read_excel("IGLOO.xls")
df.head()
```

```python
def SavetoFile(df, Sheetname):
    report_path = 'IGLOO'
    file_name = report_path +'\IGLOOOutput.xlsx'
    if not os.path.exists(report_path):
        os.makedirs(report_path)
    with pd.ExcelWriter(file_name, mode='a') as writer: 
        df.to_excel(writer, sheet_name=Sheetname, index=False)
        
df_category  
df_subCategory
df_product
df_StoreLocation
df_Store
df_Salesman
```

```python
# Adding Product Category with Primary Key
df_category = pd.DataFrame(df['ProductCategory'].unique(), columns = ['Category'])
df_category.insert(0,'CategoryID',range(1,1+df['ProductCategory'].nunique() ))
# df_category.insert(0,'CategoryID',range(1,1+len(df_category)))
# df_category['Cat_ID'] = range(1,1+df['ProductCategory'].nunique())
# df_category = df_category[['Cat_ID','Category']]
```

```python
# Adding Sub Product Category with Primary and Foreign Key
df_subCategory = df_category.merge(df[['ProductCategory','ProductSubCategory']], how='inner', 
                        left_on='Category', right_on='ProductCategory')[['CategoryID','ProductSubCategory']]
df_subCategory.drop_duplicates(inplace=True)
df_subCategory.insert(1,'SubCategoryID',range(1,1+df_subCategory['ProductSubCategory'].nunique() ))
```

```python
# df_category.merge(df[['ProductCategory','ProductSubCategory','ProductName']], how='inner', 
#                        left_on='Category', right_on='ProductCategory').merge('df_subCategory', how='inner', left_on='ProductSubCategory', right_on='ProductSubCategory')
```

```python
df_product = pd.merge(df_category.merge(df[['ProductCategory','ProductSubCategory','ProductName']], how='inner', 
                       left_on='Category', right_on='ProductCategory'), 
         df_subCategory, how='inner', left_on='ProductSubCategory', right_on='ProductSubCategory')[['CategoryID_y','SubCategoryID','ProductName']]
df_product.columns = ['CategoryID','SubCategoryID','ProductName']
df_product.drop_duplicates(inplace=True)
df_product.insert(0,'ProductID',range(1,1+len(df_product)))
```

```python
df_StoreLocation = df[['StoreRegion','StoreProvince','StoreZone','StoreArea']]
df_StoreLocation.drop_duplicates(inplace=True)
df_StoreLocation.insert(0,'StoreLocID',range(1,1+len(df_StoreLocation)))
```

```python
df_Store = df_StoreLocation.merge(df[['StoreName','StoreRegion','StoreProvince','StoreZone','StoreArea']], how='inner',
                       left_on=['StoreRegion','StoreProvince','StoreZone','StoreArea'], 
                       right_on=['StoreRegion','StoreProvince','StoreZone','StoreArea'])[['StoreLocID','StoreName']]
df_Store.drop_duplicates(inplace=True)
df_Store.insert(0,'StoreID',range(1,1+len(df_Store)))
```

```python
df_Salesman = df[['SalesMan','Route']]
df_Salesman.drop_duplicates(inplace=True)
df_Salesman.insert(0,'SalesmanID',range(1,1+len(df_Salesman)))
```
