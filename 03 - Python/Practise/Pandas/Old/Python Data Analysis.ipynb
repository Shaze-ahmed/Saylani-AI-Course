```python
import csv

moma = list(csv.reader(open("artworks_clean.csv", encoding = "utf-8")))
```

```python
header = moma[0]
moma = moma[1:]
```

```python
len(moma)
```

```python
artist = {}

for row in moma:
    art = row[1]
    if art not in artist:
        artist[art] = 1
    elif art in artist:
        artist[art] += 1
```

```python
def no_of_sub(artist_name):
    artist_value = artist[artist_name]
    template = "There are {} submission by {}"
    return template.format(artist_value,artist_name)
```
