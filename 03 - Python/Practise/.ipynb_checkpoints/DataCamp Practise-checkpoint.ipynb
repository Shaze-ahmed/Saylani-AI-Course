```python
dup_list = ['a','b','c','d','e','e','d','b','c','d','e','f','f','r','s','a','b','d','f','s','x','x','z']
dup_dict = {}
for x in dup_list:
    if x not in dup_dict:
        dup_dict[x] = 1
    else:
        dup_dict[x] += 1
        
print(dup_dict)
```

```python
from collections import Counter
type_list = Counter(dup_list)
print(type_list)
```

```python
for x, y in type_list.items():
    print(x, y)
```

```python
poker = ['Bug','Fire','Ghost','Grass','Water']
combos = []
```

```python
combos = []
for x in poker:
    for y in poker:
        if x!=y:
            if (x,y) not in combos and (y,x) not in combos:
                combos.append((x,y))
        
print(combos)
```

```python
from itertools import combinations
comb_obj = combinations(poker, 2)
combination = list(comb_obj)
print(combination)
```

```python
import numpy as np
names = ['Absol','Aron','Jynx','Natu','Onix']
attack = np.array([130, 70,50,50,45])
name_attack = zip(names, attack)
tot_attack_avg = attack.mean()
for pokeman, attacks in name_attack:    
    if attacks > tot_attack_avg:
        print("{}'s attack: {} and Average Attack is : {}" \
              .format(pokeman, attacks, tot_attack_avg))

```

```python
ylist = zip(names, attack)
xlist = []
for i in zip(names, attack):
    tlist = list(i)
    xlist.append(tlist)

maplist = [*map(list, ylist)]
maplist2 = [*map(set, zip(names, attack))]

print(xlist)
print(maplist)
print(maplist2)
```
