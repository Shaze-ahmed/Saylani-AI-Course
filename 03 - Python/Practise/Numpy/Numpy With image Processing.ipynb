```python
#  https://www.youtube.com/watch?v=C037YWKLeCE
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
## Upload image through matplolib
image = Image.open('karachi.jpg')

## converting the image into array
numpy_data = np.array(image)

numpy_data.shape
```

```python
# Display the image
plt.imshow(numpy_data)
```

```python
# Reverse the Image 
plt.imshow(numpy_data[::-1])
```

```python
# Crop the image
plt.imshow(numpy_data[1000:3000, 1000:3000])
```

```python
# Apply Filter 
plt.imshow(np.where(numpy_data > 150, 255 ,0))
```

```python
# Grayscale
plt.imshow(numpy_data[:, :, 1])
```

```python
# Landscape the Image
plt.imshow(numpy_data[:, :, 1].T)
```

```python
# GrayScale 2
R, G, B = numpy_data[:,:,0], numpy_data[:,:,1], numpy_data[:,:,2]
numpy_data = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(numpy_data)
```
