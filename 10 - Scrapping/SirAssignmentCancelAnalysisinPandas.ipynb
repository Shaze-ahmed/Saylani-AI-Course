```python
import pandas as pd
import re
from bs4 import BeautifulSoup
import requests
import csv
```

```python
csv_file = open('cancerAnalysis.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title', 'Keyword', 'Condition','Treatment','Phase'])
    
file = open("cancer_urls.txt")
link = file.readlines()
for ind, url in enumerate(link):    
    doc = ''
    l_cond = ''
    l_treat = ''
    l_phase = ''
    
    result = requests.get(url)
    doc = BeautifulSoup(result.text, "html.parser")
    
    for article in doc.find_all('div', id='main-content'):
        header = article.find('h1', class_='tr-h1 ct-sans-serif').text
        keyword = article.find('div', class_='tr-tableLabel').text
        
    for x, dd in enumerate(doc.find_all('td','ct-body3')):
        if x == 0:
            l_cond = re.sub(r"\b[\n]\b", " / ", dd.text, 2)
        elif x == 1:
            l_treat = re.sub(r"\b[\n]\b", " / ", dd.text, 2)
        elif x == 2:
            l_phase = re.sub(r"\b[\n]\b", " / ", dd.text, 2)
        else:
            break
#     print(header)
#     print(keyword)
#     print(l_cond)
#     print(l_treat)
#     print(l_phase)
    csv_writer.writerow([header, keyword, l_cond.replace("\n",""), l_treat.replace("\n",""),l_phase.replace("\n","")])
csv_file.close()
```

```python
def myCount(x):
    xx = x.split('/')
    for i in range(len(xx)):
        myval = (i+1,')',xx[i].strip())
        return myval
```
