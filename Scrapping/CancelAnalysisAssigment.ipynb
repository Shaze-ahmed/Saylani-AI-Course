```python
import pandas as pd
import re
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url1 = "https://clinicaltrials.gov/ct2/show/NCT04750434?cond=Cancer&draw=1&rank=1"
url2 = "https://clinicaltrials.gov/ct2/show/NCT04837209?cond=Cancer&draw=1&rank=2"
url3 = "https://clinicaltrials.gov/ct2/show/NCT04360330?cond=Cancer&draw=1&rank=3"
urlList = (url1, url2, url3)
```

```python
csv_file = open('cancerAnalysis.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title', 'Keyword', 'Condition','Treatment','Phase'])

for ind, i in enumerate(urlList):
    doc = ''
    l_cond = ''
    l_treat = ''
    l_phase = ''
    
    result = requests.get(i)
    doc = BeautifulSoup(result.text, "html.parser")
    
    for article in doc.find_all('div', id='main-content'):
        header = article.find('h1', class_='tr-h1 ct-sans-serif').text
        keyword = article.find('div', class_='tr-tableLabel').text
        
    for x, dd in enumerate(doc.find_all('td','ct-body3')):
        if x == 0:
            l_cond = re.sub(r"\b[\n]\b", " / ", dd.text, 2)
        elif x == 1:
            l_treat = dd.text
        elif x == 2:
            l_phase = dd.text
        else:
            break
#     print(header)
#     print(keyword)
#     print(l_cond)
#     print(l_treat)
#     print(l_phase)
    csv_writer.writerow([header, keyword, l_cond.replace("\n",""), l_treat.replace("\n",""),l_phase.replace("\n","")])
csv_file.close()
```
