```python
from bs4 import BeautifulSoup
import requests

# url = "https://clinicaltrials.gov/ct2/show/NCT04750434?cond=Cancer&draw=1&rank=1"
url = "https://clinicaltrials.gov/ct2/show/NCT04837209?cond=Cancer&draw=1&rank=2"
result = requests.get(url)
# print(result.text)
doc = BeautifulSoup(result.text, "html.parser")
```

```python
print(doc)
```

```python
headline = doc.find_all('div', id='main-content') 
# h2 = headline.find_all('div', class_='tr-results-summary')
print(headline)
```

```python
for article in doc.find_all('div', id='main-content'):
    header = article.find('h1', class_='tr-h1 ct-sans-serif').text
    print(header)
    
    keyword = article.find('div', class_='tr-tableLabel').text
    print(keyword)
    
    ds = article.find('td','ct-body3').text
    print(ds)
    
    treat = article.find('td','ct-body3')
    print(treat)
    
#     xx = treat.find()
#     outcome = article.find('div', class_='tr-indent3').text
#     print(outcome)
     
# //*[@id="tab-body"]/div/div[1]/div[2]/table/tbody/tr[2]/td[2]
```
